```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('base_de_respostas_10k_amostra.csv')
```

```python
def id_separeitor(coluna):

    df = pd.DataFrame(np.sort(coluna.unique())).reset_index()
    df = df.rename(columns={'index': 'id', 0: 'name'})
    df.id = df.id.apply(lambda x: x + 1)
    return df
```

## tabelas de identificacao

#### pais

```python
colunas = ['CompanySize',
           'CommunicationTools',
           'LanguageWorkedWith',
           'OperatingSystem',
           'Country']
empresa, ferramenta_comunic, linguagem_programacao,\
    sistema_operacional, pais = [id_separeitor(df[col]) for col in colunas]
empresa = empresa.rename({'nome': 'tamanho'})
```
